```python
##### Forest Fires #####

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fires = pd.read_csv('forestfires.csv', delimiter = ',')
fires

group_by_day = fires.groupby('day').sum()
group_by_day

group_by_month = fires.groupby('month').sum()   # the most fires occur in august and september because of
group_by_month                                  # the sum of the windspeed
windspeed = fires.groupby('month').mean()
group_by_month['windmean'] = windspeed['wind']  # the mean windspeed doesn't matter
group_by_month
group_by_month.plot(kind ='scatter', x ='wind', y='area')
group_by_month
```
